```python
import pandas as pd
import numpy as np

def move44(all_data, lst):
    req_data = [x['Close'] for x in all_data]
    req_data_ma = [x.rolling(44).mean() for x in req_data]

    slopes = []
    for series in req_data_ma:
        last_five = series.tail(5)
        x = np.arange(len(last_five))
        slope, _ = np.polyfit(x, last_five.values, 1)
        slopes.append(float(slope))

    diff = []
    close_prices = []
    ma44_values = []
    for ma, close in zip(req_data_ma, req_data):
        last_ma = float(ma.tail(1).iloc[0])
        last_close = float(close.tail(1).iloc[0])
        d = (last_ma - last_close) / last_close * 100
        diff.append(float(d))
        close_prices.append(last_close)
        ma44_values.append(last_ma)

    result = pd.DataFrame({
        'symbol': lst,
        'diff': diff,
        'slope': slopes,
        'close_price': close_prices,
        'ma44': ma44_values
    })

    print(result.dtypes)  # Debug: All should be float or int except symbol
    mask = (
        (result['diff'] > -3) &
        (result['diff'] < 3) &
        (result['slope'] > 0) &
        (result['close_price'] > result['ma44'])
    )
    masked_df = result[mask].reset_index(drop=True)
    return masked_df
```

```python
import pandas as pd
import numpy as np

def calculate_sma(prices, period):
    return prices.rolling(window=period).mean()

def calculate_ema(prices, period, sma_period):
    sma_initial = calculate_sma(prices, sma_period)  # First calculate SMA for initial EMA value
    ema = pd.Series(index=prices.index, dtype=float)
    k = 2 / (period + 1)

    # Start EMA calculation from the point where SMA first becomes available
    initial_start = sma_period + period - 1
    if initial_start < len(prices):
        ema[initial_start] = sma_initial.iloc[initial_start]  # Initialize EMA with first available SMA value

        # Calculate the rest of the EMA values
        for i in range(initial_start + 1, len(prices)):
            ema.iloc[i] = (prices.iloc[i] - ema.iloc[i-1]) * k + ema.iloc[i-1]

    return ema

def find_approaching_crossovers(all_data, symbols, threshold_percentage=0.1):
    results = []

    for data, symbol in zip(all_data, symbols):
        if 'Close' not in data.columns:
            continue

        ema_short = calculate_ema(data['Close'], 10, 5)
        ema_long = calculate_ema(data['Close'], 30, 5)

        # Calculate the percentage difference between the two EMAs in the last data point
        last_ema_diff = abs((ema_short.iloc[-1] - ema_long.iloc[-1]) / ema_long.iloc[-1]) * 100

        # Check if the EMAs are very close to each other but not crossed
        if last_ema_diff <= 0.05:
            results.append({
                'symbol': symbol,
                'EMA10': ema_short.iloc[-1],
                'EMA30': ema_long.iloc[-1],
                'percentage_diff': last_ema_diff
            })

    return pd.DataFrame(results)


approaching_crossovers = find_approaching_crossovers(all_data, nifty500, 0.1)
# print(approaching_crossovers)
```

```python
def top_gainers_and_losers(all_data, lst, days):
  req_data = [x.loc[:,['Close']] for x in all_data]
  selected_data = [x.iloc[-days:]['Close'] for x in req_data]
  change = [(x.iloc[-1] - x.iloc[0])/x.iloc[0] *100 for x in selected_data]
  result = pd.DataFrame({'symbol' :lst, 'change' :[x for x in change]}, index = [x for x in range(0,len(lst))])
  gainers = result.sort_values(by = 'change', ascending = False)
  losers = result.sort_values(by = 'change', ascending = True)
  return gainers, losers
```

```python
def move44Cross200(all_data, lst):
  req_data = [x.loc[:,['Close']] for x in all_data]
  req_data_ma44 = [x.rolling(44).mean() for x in req_data]
  req_data_ma200 = [x.rolling(200).mean() for x in req_data]

  change = [(x.tail(1).values - y.tail(1).values)/x.tail(1).values*100 for x, y in zip(req_data_ma44, req_data_ma200)]
  result = pd.DataFrame({'symbol' :lst, 'diff' :[np.abs(x[0][0]) for x in change], 'close_price': [x['Close'][-1] for x in req_data], 'ma44': [x['Close'][-1] for x in req_data_ma44], 'ma200': [x['Close'][-1] for x in req_data_ma200]},  index = [x for x in range(0,len(lst))])
  mask = (result['diff'] > -1) & (result['diff'] < 1)
  masked_df = result[mask]
  masked_df = masked_df.sort_values(by = 'diff', ascending = True)

  return masked_df
```

```python
def createZigZagPoints(dfSeries, minSegSize=0.1, sizeInDevs=0.5):
	minRetrace = minSegSize

	curVal = dfSeries[0]

	curPos = dfSeries.index[0]
	curDir = 1
	dfRes = pd.DataFrame(index=dfSeries.index, columns=["Dir", "Value"])
	for ln in dfSeries.index:
		if((dfSeries[ln] - curVal)*curDir >= 0):
			curVal = dfSeries[ln]
			curPos = ln
		else:
			retracePrc = abs((dfSeries[ln]-curVal)/curVal*100)
			if(retracePrc >= minRetrace):
				dfRes.loc[curPos, 'Value'] = curVal
				dfRes.loc[curPos, 'Dir'] = curDir
				curVal = dfSeries[ln]
				curPos = ln
				curDir = -1*curDir
	dfRes[['Value']] = dfRes[['Value']].astype(float)
	return(dfRes)
```

```python
def MagnetEffect(all_data, lst):
  req_data = [x.loc[:,['Close']] for x in all_data]
  req_data_ma44 = [x.rolling(44).mean() for x in req_data]
  req_data_ma200 = [x.rolling(200).mean() for x in req_data]



  resistance = []
  for x in all_data:
    res = createZigZagPoints(x['Close'].iloc[-30:]).dropna()
    resistance.append(max(res[res['Dir']==1]['Value'].values))



  slopes = []
  for series in req_data_ma200:
      last_five = series.tail(5)
      x = np.arange(len(last_five))
      slope, _ = np.polyfit(x, last_five, 1)
      slopes.append(slope)

  req_data = [x.loc[:,['Close']] for x in req_data]

  diff = [(x['Close'].tail(1) - y['Close'].tail(1))/x['Close'].tail(1) *100 for x, y in zip(req_data_ma44, req_data)]
  res_diff = [(x['Close'].tail(1) - y)/x['Close'].tail(1) *100 for y, x in zip(resistance, req_data)]


  result = pd.DataFrame({'symbol' :lst, 'diff' :[x.iloc[0] for x in diff],
                         'slope':[x[0] for x in slopes], 'close_price': [x['Close'][-1] for x in req_data],
                         'resistance': resistance,
                         'res_diff': [x.iloc[0] for x in res_diff],
                         'ma44': [x['Close'][-1] for x in req_data_ma44]},
                        index = [x for x in range(0,len(lst))])

  mask =  (result['slope'] > 0) & (result['close_price'] > result["ma44"]) & (result['res_diff'] > -1) & (result['res_diff'] < 1)

  masked_df = result[mask]
  return masked_df
```

```python
from datetime import date
import pandas as pd
import numpy as np
from dateutil.relativedelta import relativedelta
import yfinance as yf


def down44(lst):
  today_ = date.today()
  end_ = today_ - relativedelta(months = 4)
  end_str = end_.strftime('%Y-%m-%d')
  today_str = today_.strftime('%Y-%m-%d')
  print(today_str)

  all_data = [yf.download(f'{x}.NS', start=end_str, end=today_str) for x in lst]

  req_data = [x.loc[:,['Close']] for x in all_data]
  req_data_ma = [x.rolling(44).mean() for x in req_data]



  slopes = []
  for series in req_data_ma:
      last_five = series.tail(5)
      x = np.arange(len(last_five))
      slope, _ = np.polyfit(x, last_five, 1)
      slopes.append(slope)


  req_data = [x.loc[:,['Close']] for x in req_data]

  diff = [(x['Close'].tail(1) - y['Close'].tail(1))/x['Close'].tail(1) *100 for x, y in zip(req_data_ma, req_data)]

  result = pd.DataFrame({'symbol' :lst, 'diff' :[x.iloc[0] for x in diff], 'slope':slopes, 'close_price': [x['Close'][-1] for x in req_data]}, index = [x for x in range(0,len(lst))])

  mask = (result['diff'] > -3) & (result['diff'] < 3) & (result['slope'] < 0)
  masked_df = result[mask]
  return masked_df
```

```python
import pandas as pd
import sys
import requests
from bs4 import BeautifulSoup
import re
import os

def fetch_NSE_stock_price(stock_code):

    stock_url  = 'https://www.nseindia.com/live_market/dynaContent/live_watch/get_quote/GetQuote.jsp?symbol='+str(stock_code)
    #print(stock_url)
    headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'}
    response = requests.get(stock_url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    data_array = soup.find(id='responseDiv').getText().strip().split(":")

    for item in data_array:
        if 'lastPrice' in item:
            index = data_array.index(item)+1
            latestPrice=data_array[index].split('"')[1]
            lp =  float(latestPrice.replace(',',''))
        elif 'closePrice' in item:
            index = data_array.index(item)+1
            closePrice=data_array[index].split('"')[1]
            cp =  float(closePrice.replace(',',''))
        elif 'open' in item:
            index = data_array.index(item)+1
            openPrice=data_array[index].split('"')[1]
            op =  float(openPrice.replace(',',''))
        elif 'dayLow' in item:
            index = data_array.index(item)+1
            dayLow=data_array[index].split('"')[1]
            dl =  float(dayLow.replace(',',''))
        elif 'dayHigh' in item:
            index = data_array.index(item)+1
            dayHigh=data_array[index].split('"')[1]
            dh =  float(dayHigh.replace(',',''))
    return op,lp,dh, dl,cp

nifty50_url = 'https://www1.nseindia.com/content/indices/ind_nifty50list.csv'
df_n50 = pd.read_csv(nifty50_url)

regexp = re.compile('&')

OP  = []
LP  = []
DHP = []
DLP = []
CP  = []


while True:
   try:
      for index, row in df_n50.iterrows():
         stock_code = row['Symbol']
         if(regexp.search(stock_code) != None):
            stock_code = stock_code.replace('&','%26')

         oPrice,lPrice,dhPrice, dlPrice,cPrice = fetch_NSE_stock_price(stock_code)
         OP.append(str(oPrice))
         LP.append(str(lPrice))
         DHP.append(str(dhPrice))
         DLP.append(str(dlPrice))
         CP.append(str(cPrice))

      os.system('cls')
      print("--------------------------------------------------------------------------------------------------------------------------------------------")
      print("|{:50s} | {:20s} | {:10s} | {:10s} | {:10s} | {:10s} | {:10s}|".format( 'Company Name','Symbol','openPrice','lastPrice','dayHigh','dayLow','closePrice'))
      print("--------------------------------------------------------------------------------------------------------------------------------------------")


      for index, row in df_n50.iterrows():
         stock_code = row['Symbol']

         print("|{:50s} | {:20s} | {:10s} |{:10s} | {:10s} | {:10s} | {:10s} |".format(str(row['Company Name']), row['Symbol'], OP[index].rjust(10), LP[index].rjust(10), DHP[index].rjust(10), DLP[index].rjust(10), CP[index].rjust(10)))

      print("--------------------------------------------------------------------------------------------------------------------------------------------")

   except KeyboardInterrupt:
        sys.exit()
```
